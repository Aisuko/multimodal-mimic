# Visualization fot the 48h data with 600 epochs

```python
import re
import plotly.graph_objects as go

# Load data from file
file_path = "../train_log_p_42_600.txt"
with open(file_path, "r") as file:
    log_data = file.read()

# Parse log data
epochs = []
auc_roc_eval = []

# Regex pattern to match 'Eval' lines and capture epoch and AUC-ROC values
pattern = r"Eval:\s+Epoch\s+(\d+),\s+Loss=[\d.]+,\s+AUC-ROC=([\d.]+)"

# Find all matches in the log data
matches = re.findall(pattern, log_data)
for match in matches:
    epoch = int(match[0])
    auc_roc = float(match[1])
    epochs.append(epoch)
    auc_roc_eval.append(auc_roc)

# Create Plotly line chart
fig = go.Figure()
fig.add_trace(go.Scatter(x=epochs, y=auc_roc_eval, mode='lines+markers', name='AUC-ROC Eval'))
fig.update_layout(
    title="Epoch vs AUC-ROC (Eval)",
    xaxis_title="Epoch",
    yaxis_title="AUC-ROC",
    template="plotly_white"
)
fig.show()
```
